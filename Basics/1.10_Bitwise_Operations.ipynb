## **Bitwise Operations**


```python
import cv2
import numpy as np


def show_images_opencv(image1, image2, result_image, result_winname="Result Image"):
    cv2.imshow("Frist Image", image1)
    cv2.imshow("Second Image", image2)
    cv2.imshow(result_winname, result_image)
    if cv2.waitKey(0) == ord("q"):
        cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.style.use("grayscale")


def display_images(
    images: list, titles: list = None, rows=1, figsize=None, cmap="gray"
):
    size = len(images)
    cols = (size + rows - 1) // rows

    plt.figure(figsize=figsize)
    for i in range(size):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(images[i], cmap)
        plt.axis("off")
        if titles:
            plt.title(titles[i])

    plt.tight_layout()
    plt.show()
```

**Create Images**


```python
square = np.zeros((300, 300), dtype=np.uint8)
square = cv2.rectangle(square, (25, 25), (275, 275), 255, -1)

circle = np.zeros((300, 300), dtype="uint8")
circle = cv2.circle(circle, (150, 150), 145, 255, -1)

display_images([square, circle], ["Square", "Circle"])
```

**AND**


```python
and_image = cv2.bitwise_and(square, circle)
# show_images_opencv(square, circle, and_image, "AND")

display_images([square, circle, and_image], ["Square", "Circle", "AND"])
```

**OR**


```python
or_image = cv2.bitwise_or(square, circle)
# show_images_opencv(square, circle, or_image, "OR")

display_images([square, circle, or_image], ["Square", "Circle", "OR"])
```

**XOR**


```python
xor_image = cv2.bitwise_xor(square, circle)
# show_images_opencv(square, circle, xor_image, "XOR")

display_images([square, circle, xor_image], ["Square", "Circle", "XOR"])
```

**NOT**


```python
not_image = cv2.bitwise_not(square)
display_images([square, not_image], ["Square", "NOT"])
```

```python
not_image = cv2.bitwise_not(circle)

display_images([circle, not_image], ["Circle", "NOT"])
```

**Masking**


```python
import cv2
import numpy as np

image = cv2.imread("../Data/Images/Car.jpg")
image = cv2.resize(image, (image.shape[1] // 2, image.shape[0] // 2))
cv2.imshow("Image", image)

# Create circle mask 
mask = np.zeros(image.shape[:2], dtype=np.uint8)
center = (image.shape[1] // 2, image.shape[0] // 2)
cv2.circle(mask, center, radius=90, color=255, thickness=-1)
cv2.imshow("Mask", mask)


selected = cv2.bitwise_and(image, image, mask=mask)
cv2.imshow("Image AND Mask", selected)

if cv2.waitKey(0) == ord("q"):
    cv2.destroyAllWindows()
```
