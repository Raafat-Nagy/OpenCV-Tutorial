## **Matplotlib With OpenCV**


**How to Display a Matplotlib RGB Image**


```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("../Data/Images/Car.jpg")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis(False)
plt.show()
```

**Function for Display Images and Titles in a Dynamic Grid Layout**


```python
import matplotlib.pyplot as plt

plt.style.use("grayscale")


def display_images(
    images: list, titles: list = None, rows=1, figsize=None, cmap="gray"
):
    size = len(images)
    cols = (size + rows - 1) // rows

    plt.figure(figsize=figsize)
    for i in range(size):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(images[i], cmap)
        plt.axis("off")
        if titles:
            plt.title(titles[i])

    plt.tight_layout()
    plt.show()
```
