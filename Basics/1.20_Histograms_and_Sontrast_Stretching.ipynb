## **Image Histograms and Contrast Stretching**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### **1. Plot an Image Histograms**

```python
image = cv2.imread("../Data/Images/edges.jpg", cv2.IMREAD_GRAYSCALE)
```

**Using `cv2.calcHist`: This computes the histogram directly and plots it using matplotlib**

```python
hist = cv2.calcHist([image], [0], None, [256], [0, 256])

plt.figure(figsize=(7, 4))
plt.plot(hist, color="black")
plt.title("Grayscale Image Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

**Using `plt.hist`: This calculates and plots the histogram by flattening the image array with `.ravel()`**

```python
plt.figure(figsize=(7, 4))
plt.hist(image.ravel(), 256, [0, 256], color="gray")
plt.title("Grayscale Image Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.grid()
plt.show()
```

**Plotting Color Histograms for an Image's BGR Channels Using OpenCV**

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("../Data/Images/Car.jpg")
b_hist = cv2.calcHist([image], [0], None, [256], [0, 256])
g_hist = cv2.calcHist([image], [1], None, [256], [0, 256])
r_hist = cv2.calcHist([image], [2], None, [256], [0, 256])

plt.figure(figsize=(10, 6))
plt.plot(b_hist, color="blue", alpha=0.6, label="Blue")
plt.plot(g_hist, color="green", alpha=0.6, label="Green")
plt.plot(r_hist, color="red", alpha=0.6, label="Red")
plt.title("Color Image Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.legend()
plt.grid()
plt.show()
```

### **2. Contrast Stretching**

```python
image = cv2.imread("../Data/Images/Car.jpg",0)

# Normalize the image intensity values to the full 0255 range
min_val = np.min(image)
max_val = np.max(image)

# Perform contrast stretching
stretched = ((image - min_val) / (max_val - min_val) * 255).astype(np.uint8)


# Display original and stretched images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(stretched, cmap='gray')
plt.title('Contrast Stretched Image')
plt.axis('off')

plt.show()


stretched_hist = cv2.calcHist([stretched], [0], None, [256], [0, 256])
plt.figure(figsize=(10, 6))
plt.plot(stretched_hist, color='black')
plt.title('Histogram of Contrast Stretched Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```
