## **Otsu's Binarization**

**Doc:** https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("../Data/Images/noise.png", 0)
blured = cv2.GaussianBlur(image, (3, 3), 0)
_, thresh = cv2.threshold(blured, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)


images = [image, thresh]
titles = ["Original Image", "Thresholded Image (Otsu's Binarization)"]

plt.figure(figsize=(10, 5))
for i in range(len(images)):
    plt.subplot(1, 2, i + 1)
    plt.imshow(images[i], cmap="gray")
    plt.title(titles[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```
